### Nvidia vs Amd

```python
sns.countplot(y = 'Manufacturer', data = gpu_clean).set(title = 'Nvidia vs Amd', xlabel = 'Number of Cards')
```

#### Notes
- Majority of graphics cards developed were by Nvidia (This was due to AMD being more focused on CPU development, instead of GPU development) 
- ATI was a pioneer for early GPU development
- Intel makes onboard (integrated) GPUs, but have yet to release a full fledged consumer GPU

### Memory Speed vs Process

```python
sns.countplot(x = 'Process', hue = 'Clock Speed', data = gpu_clean, hue_order = gpu_clean['Clock Speed'].value_counts().iloc[:5].index, order = gpu_clean['Process'].value_counts().iloc[:5].index).set(title = 'Memory Speed vs Process')
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
```

#### Notes

- Majority of cards are produced with the 28nm
- Data for the newer (smaller) processes can largely be missing, due to these processes having missing data for other data columns
- General rule: Lower Process means higher power/efficiency

```python
gpu_clean['Release Date'] = gpu_clean['Release Date'].str[-5:]
```

### Memory Capacity vs Release Year

```python
sns.countplot(x = 'Release Date', hue = 'Memory', data = gpu_clean, order = gpu_clean['Release Date'].value_counts().iloc[:5].index, hue_order = gpu_clean['Memory'].value_counts().iloc[:5].index).set(title = 'Memory Capacity vs Release Year')
```

#### Notes

- As the years get closer to current day, more cards are developed with 8GB (8000 MB) of Memory. 
- Many cards in the early years were made with 2GB and people would use 2 of those cards to increase performance (SLI or Crossfire).
- SLI/Crossfire has become largely inefficient with current day GPUs (in the general consumer market)
- In current day, cards with 10GB+ Memory Capacity have been released.

### Memory Capacity vs DirectX

```python
sns.countplot(x = 'DirectX', hue = 'Memory', data = gpu_clean, order = gpu_clean['DirectX'].value_counts().iloc[:8].index, hue_order = gpu_clean['Memory'].value_counts().iloc[:5].index).set(title = 'Memory Capacity vs DirectX')
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
```

#### Notes

- Major reason why there are many cards that have a low capacity, but support DX12, is because lower end, modern cards are very low capacity. 
- Laptop GPUs are also lower capacity than desktop GPUs, which many of these could be
- DirectX 12 and 12.1 are the latest versions of DirectX (DirectX Ultimate is a branch of DX12)

#### Analysis:

As noted in my EDA notes, my research questions were designed to give me more guidance on **Who** was making a majority of GPUs and **How** much effective compute power was increased over the years. Looking through my graphs for Memory Speed, and Memory Capacity, a general trend could be observed that Memory Capacity increased as the years went by. What can be noted about this, is that GPUs with lower memory does not always equate to it being an older GPU. Many laptop (integrated) GPUs have their performance cut compared to their desktop counterparts, due to size and thermal constraints. A great example of this is the **Memory Capacity vs DirectX** graph; DirectX 12 is currently the most recent version of DirectX (To reiterate, DX Ultimate is a branch of DX12), and many lower end GPUs from previous generations can support the use of DX12. Nvidia graphics cards from the 1000 series can even support RTX (without DLSS). The 4GB capacity GPUs are what were considered *budget* GPUs back in 2016, where 8GB+ were usually at the upper-mid to high-end GPUs. Looking at the **Memory Speed vs Process** graph, we can observe that having a smaller process (gpu die {processing chip} size) generally resulted in higher clock (MHz) speed. A smaller process means that the amount of performance lost to heat is decreased by a sizeable factor. This directly leads to an increase in memory speed, since GPUs (this is more of an Nvidia trend, but can be applied overall) will keep pushing the clock speed higher, until it hits a wall due to heat constraints. Sub-Zero and liquid cooled GPUs have become readily more popular over the years due to the fact that increased thermal dissipation directly contributes to increased performance. One major **important** note, is that there are missing data samples, due to certain GPUs not having enough information for me to consider it valid data to use for analysis.
