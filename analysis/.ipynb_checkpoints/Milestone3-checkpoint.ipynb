### More Data Cleaning

#### Important Notes About the Data
- Clock Speed is in MHz
- Bandwidth is in GB/s
- Memory is in MB
- Memory Bus is in Bit
- Cache is in KB
- Power Consumption is in W
- Pixel Speed is in GPixel/s
- Process is in nm
- Release Price is in Dollars
- Texture Rate is in GTexel/s

```python
import pandas as pd

clean2 = pd.read_csv("../data/processed/Cleaned_GPU_df.csv")
clean2
```

```python
clean2.replace({' MHz': ''}, regex = True, inplace = True)
clean2
```

```python
clean2['Clock Speed'] = pd.to_numeric(clean2['Clock Speed'])
clean2
```

```python
clean2.replace({'GB/sec': '', 'KB': '', 'KB\(x2\)': '', ' Watts': '', ' Bit': '', ' GPixel/s': '', 'nm': '', '$': '', '$ ': '', 'GTexel/s': '', ' MB': ''}, regex = True, inplace = True)
clean2
```

```python
clean2.replace({' W': ''}, regex = True, inplace = True)
clean2
```

```python
clean2.replace({'0\(x2\)': ''}, regex = True, inplace = True)
```

```python
#clean2['Cache'] = pd.to_numeric(clean2['Cache'])
clean2['Power Consumption'] = pd.to_numeric(clean2['Power Consumption'])
clean2['Bandwidth'] = pd.to_numeric(clean2['Memory'])
clean2['Memory Bus'] = pd.to_numeric(clean2['Memory Bus'])
clean2['Pixel Speed'] = pd.to_numeric(clean2['Pixel Speed'])
clean2['Process'] = pd.to_numeric(clean2['Process'])
clean2['Texture Rate'] = pd.to_numeric(clean2['Texture Rate'])
clean2
```

```python
clean2.to_csv(r'C:\Users\rabid\Desktop\Important Files\UNI\UBCO 3 Term 2\COSC 301\301 Project\course-project-solo_109\data\processed\Final_Clean_GPU_df.csv', index = False)
```
